Process Query File

```python
# 1) Download the query tweets from the Gitlab repository: https://gitlab.com/checkthat_lab/clef2025-checkthat-lab/-/tree/main/task4?ref_type=heads
# 2) Drag and drop the downloaded file to the "Files" section (left vertical menu on Colab)
# 3) Modify the path to your local file path
PATH_QUERY_DATA = '/home/guilherme/workspace/feup-ir-project/data/subtask_4b/subtask4b_query_tweets.tsv' #MODIFY PATH
df_query = pd.read_csv(PATH_QUERY_DATA, sep = '\t')
df_query.head()
```

```python
df_query.info()
```

```python
from sklearn.model_selection import train_test_split

TRAINING_SPLIT_SIZE = 0.8

df_training_set, df_test_set = train_test_split(df_query, test_size=0.2)

print("--- Training")
df_training_set.info()
print("--- Test")
df_test_set.info()
```

```python
df_training_set.head()
```

```python
df_collection.head()
```

```python
positive_train_data = pd.DataFrame(columns=df_collection.columns)
query = []
for idx, row in df_training_set.iterrows():
    entry_cord_id = row.cord_uid
    query.append(row.tweet_text)
    positive_train_data = pd.concat([positive_train_data, df_collection.loc[df_collection["cord_uid"] == entry_cord_id]], ignore_index=True)

cls = [1] * len(positive_train_data)
positive_train_data["class"] = cls
positive_train_data["query"] = query
positive_train_data.info()
```

```python
positive_test_data = pd.DataFrame(columns=df_collection.columns)
query = []
for idx, row in df_test_set.iterrows():
    entry_cord_id = row.cord_uid
    query.append(row.tweet_text)
    positive_test_data = pd.concat([positive_test_data, df_collection.loc[df_collection["cord_uid"] == entry_cord_id]], ignore_index=True)

cls = [1] * len(positive_test_data)
positive_test_data["class"] = cls
positive_test_data["query"] = query
positive_test_data.info()
```

```python
NUM_NEGATIVE_SAMPLES = 5

negative_train_data = pd.DataFrame(columns=df_collection.columns)
query = []

for idx, row in df_training_set.iterrows():
    entry_cord_id = row.cord_uid
    repeated_queries = [row.tweet_text] * NUM_NEGATIVE_SAMPLES
    query += repeated_queries
    negative_samples = df_collection.loc[df_collection["cord_uid"] != entry_cord_id]
    negative_samples = negative_samples.sample(NUM_NEGATIVE_SAMPLES)
    negative_train_data = pd.concat([negative_train_data, negative_samples], ignore_index=True)

cls = [0] * len(negative_train_data)
negative_train_data["class"] = cls
negative_train_data["query"] = query
negative_train_data.head()
```

```python
NUM_NEGATIVE_SAMPLES = 5

negative_test_data = pd.DataFrame(columns=df_collection.columns)
query = []

for idx, row in df_test_set.iterrows():
    entry_cord_id = row.cord_uid
    repeated_queries = [row.tweet_text] * NUM_NEGATIVE_SAMPLES
    query += repeated_queries
    negative_samples = df_collection.loc[df_collection["cord_uid"] != entry_cord_id]
    negative_samples = negative_samples.sample(NUM_NEGATIVE_SAMPLES)
    negative_test_data = pd.concat([negative_test_data, negative_samples], ignore_index=True)

cls = [0] * len(negative_test_data)
negative_test_data["class"] = cls
negative_test_data["query"] = query
negative_test_data.head()
```

```python
mixed_training_set = pd.concat([negative_train_data, positive_train_data], ignore_index=True)
mixed_training_set.info()

mixed_training_set.to_csv('data/train/tweets-query-pairs.tsv',sep='\t')
```

```python
mixed_test_set = pd.concat([negative_test_data, positive_test_data], ignore_index=True)
mixed_test_set.info()

mixed_test_set.to_csv('data/test/tweets-query-pairs.tsv',sep='\t')
```
